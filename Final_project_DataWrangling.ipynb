###1. Import the dataset using Pandas from above mentioned url.

3. Low Level Data Understanding :


> a. Find count of unique values in location column.

> b. Find which continent has maximum frequency using values
counts.

> c. Find maximum & mean value in 'total_cases'.

> d. Find 25%,50% & 75% quartile value in 'total_deaths'.

> e. Find which continent has maximum
'human_development_index'.

> f. Find which continent has minimum 'gdp_per_capita'.



```python
#Find count of unique values in location column

len(data["location"].unique())
```

```python
# Find which continent has maximum frequency using values counts
data["continent"].value_counts().idxmax()
```

```python
#Find maximum and mean value in 'total cases'
maximum_cases = data['total_cases'].max()
mean_cases = data['total_cases'].mean()
print("Maximum Cases : ", maximum_cases)
print("Mean Cases : ", round(mean_cases,2))
```

```python
#Find 25%, 50% and 75% quartile value in 'total_deaths'
data['total_deaths'].quantile([0.25,0.5,0.75])
```

```python
# Find which continent has maximum 'human_development_index'.
data.loc[data['human_development_index']==data['human_development_index'].max(),"continent"].unique()[0]
```

```python
#Find which continent has minimum 'gdp_per_capita'
data.loc[data['gdp_per_capita']==data['gdp_per_capita'].min(),"continent"].unique()[0]
```

###4. Filter the dataframe with only this columns
['continent','location','date','total_cases','total_deaths','gdp_per_ca
pita','
human_development_index'] 
and update the data frame.


```python
data = data[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
data.head()
```

###5. Data Cleaning
 a. Remove all duplicates observations

 b. Find missing values in all columns

 c. Remove all observations where continent column value is 

 d. Fill all missing values with 0


```python
#Remove all duplicate observations
data.drop_duplicates(inplace=True)
```

```python
#Find missing values in all columns
data.isna().sum()
```

```python
#Remove all observations where continent column value is missing
data.dropna(subset=['continent'],inplace=True)
data.shape
```

```python
# Fill all missing values with 0
data.fillna(0)
```

###6. Date time format :
 a. Convert date column in datetime format using
pandas.to_datetime

 b. Create new column month after extracting month data from
date
 column.

```python
# Convert date column in datetime format using pandas.to_datetime
data['date']= pd.to_datetime(data['date'])
data['date']
```

```python
# Create new column month after extracting month data from date column.
data['month'] = [d.strftime("%B") for d in data['date']]
data['month']
```

###7. Data Aggregation:
 a. Find max value in all columns using groupby function on
'continent'
 column
 Tip: use reset_index() after applying groupby
 
 b. Store the result in a new dataframe named 'df_groupby'.
 (Use df_groupby dataframe for all further analysis)

```python
#Find max value in all columns using groupby function on 'continent' column Tip: use reset_index() after applying groupby
df_groupby = data.groupby('continent').max().reset_index()
df_groupby
```

###8. Feature Engineering :
 a. Create a new feature 'total_deaths_to_total_cases' by ratio of
 'total_deaths' column to 'total_cases'


```python
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths']/df_groupby['total_cases']
df_groupby
```

###9. Data Visualization :
 a. Perform Univariate analysis on 'gdp_per_capita' column by
plotting
 histogram using seaborn dist plot.

 b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'

 c. Plot Pairplot on df_groupby dataset.

```python
import matplotlib as plt
import seaborn as sns
```

```python
#a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.
sns.displot(df_groupby['gdp_per_capita'], kde=False,color="RED")
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_groupby['gdp_per_capita'], df_groupby['total_cases'])
plt.xlabel('GDP per capita')
plt.ylabel('Total cases')
plt.show()
```

```python
#c. Plot Pairplot on df_groupby dataset.
sns.pairplot(df_groupby)
```

```python
#d. Plot a bar plot of 'continent' column with 'total_cases' . Tip : using kind='bar' in seaborn catplot
sns.catplot(x='continent', y='total_cases', kind='bar', data=df_groupby)
```
